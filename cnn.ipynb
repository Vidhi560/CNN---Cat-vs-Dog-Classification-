```python
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
#Dense = activation layer 
#Conv2D =image data convert and compress
#Maxpooling2D = 
#Flatten =convert 2D data to 1D data
#sequential = add all this in sequense with conaction 
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python

cnn = Sequential()
```

```python
cnn.add(Conv2D(32,(3,3),input_shape = (64,64,3),activation = "relu"))# bit size (image size fix)=32,(64,64,3)for colord and (64,64)for black
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Conv2D(32,(3,3),activation = "relu"))
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Flatten())
```

```python
cnn.add(Dense(64,activation = "relu"))
cnn.add(Dense(32,activation = "relu"))
cnn.add(Dense(16,activation = "relu"))
cnn.add(Dense(8,activation = "relu"))
cnn.add(Dense(4,activation = "relu"))
cnn.add(Dense(1,activation = "sigmoid"))
```

```python
cnn.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        r"C:\Users\Vidhi Prajapati\OneDrive\Desktop\cnn cat dog dataset test and train\training_set\training_set",
        target_size=(64,64),
        batch_size=32,
        class_mode='binary')

test_generator = test_datagen.flow_from_directory(
        r"C:\Users\Vidhi Prajapati\OneDrive\Desktop\cnn cat dog dataset test and train\test_set\test_set",
        target_size=(64,64),
        batch_size=32,
        class_mode='binary')

cnn.fit(
        train_generator,
        steps_per_epoch=2000,
        epochs=50,
        validation_data=test_generator)
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np
img = image.load_img(r"C:\Users\Vidhi Prajapati\OneDrive\Desktop\cnn cat dog dataset test and train\training_set\training_set\dogs\dog.50.jpg")
img = image.img_to_array(img)
img = np.expand_dims(img, axis=0)
img = img / 255.0
result = cnn.predict(img)
```

```python
if result[0][0] > 0.5:
    print("CAT")
else: 
    print("DOG")
```
